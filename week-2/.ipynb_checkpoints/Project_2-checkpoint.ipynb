```python
## Cubic Equation Program
import numpy as np
while True: #loop to allow repeated calculations
    print("EQUATION OF ROOTS")

    print("1. Quadratic Equation")
    print("2. Cubic Equation")
    print("3. Quartic Equation")

    choice = int(input("Choose an Option (1-3):"))
    if choice == 1:
        print("Quadratic Equation: ax^2 + bx + c= 0")
        a = float(input("Enter a: "))
        b = float(input("Enter b: "))
        c = float(input("Enter c: "))
    
        roots = np.roots([a,b,c,])
        print("The roots of the equation are: ", np.round(roots, 4))

    elif choice == 2:
        print("Cubic Equation: ax^3 + bx^2 + cx + d = 0")
        a = float(input("Enter a: "))
        b = float(input("Enter b: "))
        c = float(input("Enter c: "))
        d = float(input("Enter d: "))

        roots = np.roots([a,b,c,d])
        print("The roots of the equation are: ", np.round(roots, 4))

    elif choice == 3:
         print("Quartic Equation: ax^4 + bx^3 + cx^2 + dx + e = 0")
         a = float(input("Enter a: "))
         b = float(input("Enter b: "))
         c = float(input("Enter c: "))
         d = float(input("Enter d: "))
         e = float(input("Enter e: "))

         roots = np.roots([a,b,c,d,e])
         print("The roots of the equation are: ", np.round(roots, 4))

    else:
        print("Invalid choice")

    repeat = input("Do you want to calculate again? Yes / No: ").strip().lower()
    if repeat != "yes":
        print("Program exited Successfully")
        break
    
```
