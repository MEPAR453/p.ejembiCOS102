```python
import tkinter as tk
from tkinter import messagebox as msgbox
from tkinter import ttk
import pandas as pd

#load and read csv file
df =  pd.read_csv("GIG-logistics.csv")

#Get Departments
departments = sorted (df["Department"].unique())

#function to check employee
def verify():
    name = name_entry.get().strip().lower()
    dept = dept_combo.get().strip().lower()
    
    found = False
    department_members = []

    for index, row in df.iterrows():
        if row["Name"].lower() == name and row["Department"].lower() == dept:
            found = True
            employee_id = row["EmployeeID"]
            employee_role = row["Role"]
            employee_email = row["Email"]

        if row["Department"].lower() == dept:
            department_members.append(row["Name"])

    #Clear previous result
    result_text.delete(1.0, tk.END)
    other_members_text.delete(1.0, tk.END)
    
    if found:
        result_text.insert(tk.END, f"Welcome {name.title()}!\n")
        result_text.insert(tk.END, f"EmployeeID: {employee_id}\n")
        result_text.insert(tk.END, f"Role: {employee_role}\n")
        result_text.insert(tk.END, f"Email: {employee_email}\n\n")
       
        for member in department_members:
            if member.lower() != name:
                other_members_text.insert(tk.END, f"- {member}\n")

    else:
        result_text.insert(tk.END, "Employee does not exist.\n")

def clear():
    name_entry.delete(0, tk.END)
    dept_combo.set("")
    result_text.delete(1.0, tk.END) 

#Create Window
root = tk.Tk()
root.title("GIG Logistics Employee Verification")
root.geometry("400x400")
root.resizable(True,True)

#Labels and Entries
# Header at the top
header_label = tk.Label(root, text="GIG Logistics",
                        font=("Helvetica", 16, "bold"), fg="white", bg="dark blue", pady=10)
header_label.pack(fill="x")
header_label = tk.Label(root, text="Employee verification system",
                        font=("Helvetica", 10,"bold"), fg="white", bg="black", pady=10)
header_label.pack(fill="x")

# Big frame for inputs
# Title above frame
tk.Label(root, text="Employee Verification",
         font=("Arial", 8, "bold")).pack(anchor="w", padx=10)

# Normal frame (no text)
big_frame = tk.Frame(root, padx=10, pady=10, relief="groove", bd=2)
big_frame.pack(padx=10, pady=5, fill="x")

tk.Label(big_frame , text="Enter Name:").grid(row=0, column=0, sticky="w", pady=5)
name_entry = tk.Entry(big_frame, width=30)
name_entry.grid(row=0, column=1, pady=5, padx=5)

tk.Label(big_frame , text="Select Department:").grid(row=1, column=0, sticky="w", pady=5)
dept_combo = ttk.Combobox(big_frame, values=departments, width=28, state="readonly")
dept_combo.grid(row=1, column=1, pady=5, padx=5)

# Button
button_frame = tk.Frame(root)
button_frame.pack(pady=10)

tk.Button(button_frame, text="Verify", command=verify, bg="blue", fg="white").grid(row=0, column=0, padx=10)
tk.Button(button_frame, text="Clear", command=clear, bg="red", fg="white").grid(row=0, column=1, padx=10)

# Output box
# Result Frame
tk.Label(root, text="Result",
         font=("Arial", 8, "bold")).pack(anchor="w", padx=10)

result_frame = tk.Frame(root, padx=10, pady=10, relief="groove", bd=2)
result_frame.pack(padx=10, pady=5, fill="both", expand=True)

# Employee Info Text
result_text = tk.Text(result_frame, height=8, width=60)
result_text.pack(padx=5, pady=5, fill="x")

# Nested frame for Other Members
other_members_frame = tk.LabelFrame(result_frame, text="Other Members in Department", padx=10, pady=10)
other_members_frame.pack(padx=5, pady=5, fill="both", expand=True)

# Text box for other members
other_members_text = tk.Text(other_members_frame, height=6, width=60, fg="green")
other_members_text.pack(padx=5, pady=5, fill="both", expand=True)

# Run app
root.mainloop()

```
