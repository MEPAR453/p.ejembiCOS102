```python
import tkinter as tk
from tkinter import messagebox as msgbox
from tkinter import ttk
import pandas as pd

#load and read csv file
df =  pd.read_csv("GIG-logistics.csv")

#Get Departments
departments = sorted (df["Department"].unique())

#function to check employee
def verify():
    name = name_entry.get().strip().lower()
    dept = dept_combo.get().strip().lower()
    
    found = False
    department_members = []

    for index, row in df.iterrows():
        if row["Name"].lower() == name and row["Department"].lower() == dept:
            found = True
            employee_id = row["EmployeeID"]
            employee_role = row["Role"]
            employee_email = row["Email"]

        if row["Department"].lower() == dept:
            department_members.append(row["Name"])

    #Clear previous result
    result_text.delete(1.0, tk.END)
    other_members_text.delete(1.0, tk.END)
    
    if found:
        result_text.insert(tk.END, f"Welcome {name.title()}!\n")
        result_text.insert(tk.END, f"EmployeeID: {employee_id}\n")
        result_text.insert(tk.END, f"Role: {employee_role}\n")
        result_text.insert(tk.END, f"Email: {employee_email}\n\n")
       
        for member in department_members:
            if member.lower() != name:
                other_members_text.insert(tk.END, f"- {member}\n")

    else:
        msgbox.showinfo("Result", "Employee does not exist.")

def clear():
    name_entry.delete(0, tk.END)
    dept_combo.set("")
    result_text.delete(1.0, tk.END) 

#Create Window
root = tk.Tk()
root.title("GIG Logistics Employee Verification")
root.geometry("400x400")
root.resizable(True,True)

#Labels and Entries
# Header at the top
header_label = tk.Label(root, text="GIG Logistics",
                        font=("Helvetica", 16, "bold"), fg="white", bg="dark blue", pady=10)
header_label.pack(fill="x")
header_label = tk.Label(root, text="Employee verification system",
                        font=("Helvetica", 8,"bold"), fg="white", bg="black", pady=10)
header_label.pack(fill="x")
# Big frame for inputs
big_frame = tk.LabelFrame(root, text="Employee Verification", padx=10, pady=10)
big_frame.pack(padx=10, pady=10, fill="x")

tk.Label(big_frame , text="Enter Name:").grid(row=0, column=0, sticky="w", pady=5)
name_entry = tk.Entry(big_frame, width=30)
name_entry.grid(row=0, column=1, pady=5, padx=5)

tk.Label(big_frame , text="Select Department:").grid(row=1, column=0, sticky="w", pady=5)
dept_combo = ttk.Combobox(big_frame, values=departments, width=28, state="readonly")
dept_combo.grid(row=1, column=1, pady=5, padx=5)

# Button
button_frame = tk.Frame(root)
button_frame.pack(pady=10)

tk.Button(button_frame, text="Verify", command=verify, bg="blue", fg="white").grid(row=0, column=0, sticky="w", padx=50)
tk.Button(button_frame, text="Clear", command=clear, bg="red", fg="white").grid(row=0, column=1, sticky="e", padx=50)

# Output box
# Result Frame
result_frame = tk.LabelFrame(root, text="Result", padx=10, pady=10)
result_frame.pack(padx=10, pady=10, fill="both", expand=True)

# Employee Info Text
result_text = tk.Text(result_frame, height=8, width=60)
result_text.pack(padx=5, pady=5, fill="x")

# Nested frame for Other Members
other_members_frame = tk.LabelFrame(result_frame, text="Other Members in Department", padx=10, pady=10)
other_members_frame.pack(padx=5, pady=5, fill="both", expand=True)

# Text box for other members
other_members_text = tk.Text(other_members_frame, height=6, width=60, fg="green")
other_members_text.pack(padx=5, pady=5, fill="both", expand=True)

# Run app
root.mainloop()

```

```python
import tkinter as tk
from tkinter import ttk, messagebox
import pandas as pd
from tkinter import scrolledtext

class EmployeeVerificationSystem:
    def __init__(self, root):
        self.root = root
        self.root.title("GIG Logistics Employee Verification System")
        self.root.geometry("800x650")
        self.root.resizable(True, True)
        
        # Load employee data
        try:
            self.df = pd.read_csv('GIG-logistics.csv')
            print("CSV Columns:", self.df.columns.tolist())  # Debug line
        except FileNotFoundError:
            messagebox.showerror("Error", "GIG-logistics.csv file not found!")
            self.df = pd.DataFrame()
        except Exception as e:
            messagebox.showerror("Error", f"Error loading CSV: {str(e)}")
            self.df = pd.DataFrame()
        
        # Style
        style = ttk.Style()
        style.theme_use('clam')
        
        # Configure colors
        self.root.configure(bg='#f0f0f0')
        
        # Create main frame
        main_frame = ttk.Frame(root, padding="20")
        main_frame.pack(fill=tk.BOTH, expand=True)
        
        # Title with company logo effect
        title_frame = ttk.Frame(main_frame)
        title_frame.pack(fill=tk.X, pady=(0, 20))
        
        title_label = ttk.Label(title_frame, text="GIG LOGISTICS", 
                                font=('Arial', 24, 'bold'), foreground='#2c3e50')
        title_label.pack()
        
        subtitle_label = ttk.Label(title_frame, text="Employee Verification System", 
                                  font=('Arial', 12), foreground='#7f8c8d')
        subtitle_label.pack()
        
        # Input Frame
        input_frame = ttk.LabelFrame(main_frame, text="Employee Information", padding="20")
        input_frame.pack(fill=tk.X, pady=10)
        
        # Name input
        ttk.Label(input_frame, text="Full Name:", font=('Arial', 11)).grid(row=0, column=0, sticky=tk.W, pady=8)
        self.name_entry = ttk.Entry(input_frame, width=40, font=('Arial', 11))
        self.name_entry.grid(row=0, column=1, padx=15, pady=8, sticky=tk.W)
        
        # Department input with combobox for easy selection
        ttk.Label(input_frame, text="Department:", font=('Arial', 11)).grid(row=1, column=0, sticky=tk.W, pady=8)
        
        # Get unique departments for dropdown
        if not self.df.empty and 'Department' in self.df.columns:
            departments = sorted(self.df['Department'].unique())
            self.dept_var = tk.StringVar()
            self.dept_combo = ttk.Combobox(input_frame, textvariable=self.dept_var, 
                                          values=departments, width=38, font=('Arial', 11))
            self.dept_combo.grid(row=1, column=1, padx=15, pady=8, sticky=tk.W)
            self.dept_combo.set('')  # Empty default
        else:
            self.dept_entry = ttk.Entry(input_frame, width=40, font=('Arial', 11))
            self.dept_entry.grid(row=1, column=1, padx=15, pady=8, sticky=tk.W)
        
        # Buttons Frame
        button_frame = ttk.Frame(input_frame)
        button_frame.grid(row=2, column=0, columnspan=2, pady=20)
        
        # Verify button with style
        self.verify_btn = tk.Button(button_frame, text="🔍 Verify Employee", 
                                   command=self.verify_employee, 
                                   bg='#27ae60', fg='white', font=('Arial', 11, 'bold'),
                                   padx=20, pady=8, cursor='hand2')
        self.verify_btn.pack(side=tk.LEFT, padx=10)
        
        # Clear button
        self.clear_btn = tk.Button(button_frame, text="🗑️ Clear Fields", 
                                  command=self.clear_fields,
                                  bg='#e74c3c', fg='white', font=('Arial', 11, 'bold'),
                                  padx=20, pady=8, cursor='hand2')
        self.clear_btn.pack(side=tk.LEFT, padx=10)
        
        # Results Frame
        results_frame = ttk.LabelFrame(main_frame, text="Verification Results", padding="15")
        results_frame.pack(fill=tk.BOTH, expand=True, pady=10)
        
        # Result message with custom styling
        self.result_frame = tk.Frame(results_frame, bg='white', relief=tk.RIDGE, bd=2)
        self.result_frame.pack(fill=tk.X, pady=(0, 10))
        
        self.result_label = tk.Label(self.result_frame, text="", font=('Arial', 12, 'bold'),
                                     bg='white', pady=10)
        self.result_label.pack()
        
        # Department members display
        members_header_frame = tk.Frame(results_frame, bg='#3498db')
        members_header_frame.pack(fill=tk.X, pady=(5, 0))
        
        tk.Label(members_header_frame, text="👥 Department Members", 
                font=('Arial', 11, 'bold'), bg='#3498db', fg='white', pady=5).pack()
        
        # Scrolled text for department members
        self.members_text = scrolledtext.ScrolledText(results_frame, width=70, height=12, 
                                                      font=('Courier', 10), wrap=tk.WORD,
                                                      bg='white', relief=tk.SUNKEN, bd=2)
        self.members_text.pack(fill=tk.BOTH, expand=True, pady=5)
        
        # Configure text tags for formatting
        self.members_text.tag_configure('header', font=('Arial', 11, 'bold'), foreground='#2c3e50')
        self.members_text.tag_configure('member', font=('Arial', 10), foreground='#34495e')
        self.members_text.tag_configure('suggestion', font=('Arial', 10, 'italic'), foreground='#e67e22')
        
        # Statistics Frame
        stats_frame = ttk.LabelFrame(main_frame, text="Company Overview", padding="10")
        stats_frame.pack(fill=tk.X, pady=10)
        
        # Load statistics
        if not self.df.empty:
            total_employees = len(self.df)
            departments = self.df['Department'].nunique()
            
            stats_inner = tk.Frame(stats_frame, bg='#ecf0f1')
            stats_inner.pack(fill=tk.X, padx=5, pady=5)
            
            tk.Label(stats_inner, text=f"👥 Total Employees: {total_employees}", 
                    font=('Arial', 10, 'bold'), bg='#ecf0f1', fg='#2980b9').pack(side=tk.LEFT, padx=30)
            tk.Label(stats_inner, text=f"🏢 Departments: {departments}", 
                    font=('Arial', 10, 'bold'), bg='#ecf0f1', fg='#2980b9').pack(side=tk.LEFT, padx=30)
        
        # Bind Enter key to verify function
        self.root.bind('<Return>', lambda event: self.verify_employee())
        
        # Focus on name entry
        self.name_entry.focus()
        
    def verify_employee(self):
        """Verify if the employee exists in the database"""
        # Clear previous results
        self.members_text.delete(1.0, tk.END)
        self.result_label.config(text="")
        
        # Get input values
        name = self.name_entry.get().strip()
        if hasattr(self, 'dept_combo'):
            department = self.dept_combo.get().strip()
        else:
            department = self.dept_entry.get().strip()
        
        # Validate input
        if not name or not department:
            messagebox.showwarning("Input Error", "Please enter both name and department!")
            return
        
        if self.df.empty:
            messagebox.showerror("Error", "No employee data available!")
            return
        
        # Check if employee exists (case-insensitive)
        name_lower = name.lower()
        dept_lower = department.lower()
        
        employee_exists = False
        employee_data = None
        
        for idx, row in self.df.iterrows():
            if (str(row['Name']).strip().lower() == name_lower and 
                str(row['Department']).strip().lower() == dept_lower):
                employee_exists = True
                employee_data = row
                break
        
        # Display result
        if employee_exists:
            # Format welcome message with employee details
            welcome_msg = f"✅ WELCOME, {employee_data['Name']}!\n"
            welcome_msg += f"   Department: {employee_data['Department']}\n"
            welcome_msg += f"   Role: {employee_data['Role']}\n"
            welcome_msg += f"   Location: {employee_data['Location']}"
            
            self.result_label.config(text=welcome_msg, fg='#27ae60')
            
            # Get and display all members from the same department
            dept_members = self.df[self.df['Department'].str.lower() == dept_lower]
            
            if not dept_members.empty:
                # Header
                self.members_text.insert(tk.END, f"Members in {department.title()} Department:\n", 'header')
                self.members_text.insert(tk.END, "=" * 50 + "\n\n", 'header')
                
                # List all members with their roles
                for idx, member in dept_members.iterrows():
                    member_info = f"• {member['Name']}\n"
                    member_info += f"  Role: {member['Role']}\n"
                    member_info += f"  Location: {member['Location']}\n"
                    member_info += f"  Email: {member['Email']}\n\n"
                    self.members_text.insert(tk.END, member_info, 'member')
        else:
            self.result_label.config(text="❌ EMPLOYEE NOT FOUND in our records!", fg='#e74c3c')
            
            # Suggest similar names if any
            similar_employees = self.find_similar_employees(name)
            if similar_employees:
                self.members_text.insert(tk.END, "Did you mean one of these employees?\n\n", 'suggestion')
                for emp_name, emp_dept in similar_employees:
                    self.members_text.insert(tk.END, f"• {emp_name} ({emp_dept})\n", 'suggestion')
            
            # Show all employees in the department if department exists
            if not self.df.empty and 'Department' in self.df.columns:
                dept_exists = any(str(dept).lower() == dept_lower for dept in self.df['Department'].unique())
                if dept_exists:
                    self.members_text.insert(tk.END, f"\nEmployees in {department.title()} department:\n\n", 'header')
                    dept_employees = self.df[self.df['Department'].str.lower() == dept_lower]
                    for idx, emp in dept_employees.iterrows():
                        self.members_text.insert(tk.END, f"• {emp['Name']}\n", 'member')
    
    def find_similar_employees(self, search_name):
        """Find employees with similar names"""
        if self.df.empty:
            return []
        
        similar = []
        search_parts = search_name.lower().split()
        
        for idx, row in self.df.iterrows():
            name = str(row['Name'])
            name_lower = name.lower()
            
            # Check for partial matches
            if any(part in name_lower for part in search_parts):
                similar.append((name, row['Department']))
        
        # Return top 5 unique matches
        seen = set()
        unique_similar = []
        for emp in similar:
            if emp[0] not in seen:
                seen.add(emp[0])
                unique_similar.append(emp)
                if len(unique_similar) >= 5:
                    break
        
        return unique_similar
    
    def clear_fields(self):
        """Clear all input fields and results"""
        self.name_entry.delete(0, tk.END)
        if hasattr(self, 'dept_combo'):
            self.dept_combo.set('')
        else:
            self.dept_entry.delete(0, tk.END)
        self.result_label.config(text="")
        self.members_text.delete(1.0, tk.END)
        self.name_entry.focus()

def main():
    root = tk.Tk()
    app = EmployeeVerificationSystem(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
