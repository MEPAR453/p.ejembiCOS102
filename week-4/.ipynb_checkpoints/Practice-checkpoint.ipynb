```python
import pandas as pd
df = pd.read_csv("diabetes.csv")
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
data = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
for col in data:
    df[col] = df[col].fillna(df[col].mean())
df
```

```python
data = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
for col in data:
    df[col] = df[col].fillna(df[col].median())
df
```

```python
data = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
for col in data:
    df[col] = df[col].fillna(df[col].mode())
df
```
