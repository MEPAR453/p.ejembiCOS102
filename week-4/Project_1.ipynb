```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. LOAD THE DATA (This is the missing step!)
df = pd.read_csv('New York City Airbnb Open Data.csv') 

plt.figure(figsize=(10, 6))
sns.barplot(x='room_type', y='price', data=df, palette='viridis', ci=None)
plt.title('Average Price per Room Type')
plt.ylabel('Average Price ($)')
plt.show()

plt.savefig('chart_name.png')
```

Insight 1: Based on the visualization, it is evident that room type is a primary determinant of pricing within the New York City Airbnb ecosystem. This data indicates that the NYC market is heavily skewed toward high-value, independent stays, suggesting that property owners who offer entire units stand to gain the highest return on investment compared to those offering communal or shared accommodations.

Entire homes/apartments command the highest average price, roughly double that of private rooms. This suggests that privacy is a major price driver in the NYC market.

```python
plt.figure(figsize=(12, 8))
# We filter prices under $500 to keep the chart readable (removing extreme outliers)
sns.boxplot(x='neighbourhood_group', y='price', data=df[df.price < 500])
plt.title('Density and Distribution of Prices by Borough')
plt.show()
plt.savefig('chart_name.png')
```

Insight 2: The data confirms a clear geographical hierarchy in NYC Airbnb pricing. Manhattan sits in a league of its own as the premium market, while Brooklyn serves as a mid-tier bridge. The Bronx, Queens, and Staten Island offer a highly consistent budget-friendly alternative. 

Manhattan and Brooklyn have the widest price range and the highest density of listings. Manhattan is the most expensive borough, likely due to its proximity to major tourist attractions.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='availability_365', y='number_of_reviews', data=df, alpha=0.5)
plt.title('Relationship between Availability and Popularity')
plt.show()
plt.savefig('chart_name.png')
```

Insight 3: The scatter plot reveals that popularity measured by the number of reviews is not a simple byproduct of being often available. 
There is a high concentration of listings with zero availability that also have high review counts. This indicates a group of 'highly successful' listings that are almost constantly booked.
