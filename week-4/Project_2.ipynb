```python
import pandas as pd
df = pd.read_csv("Titanic.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
data = ["Age"]
for col in data:
    df[col] = df[col].fillna(df[col].mean())
df
```

```python
# Find the most frequent value (Mode)
mode_value = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(mode_value)
df
```

```python
df.isna().sum()
```

```python
df['Cabin'] = df['Cabin'].fillna('Unknown')
df
```

```python
df.isna().sum()
```

```python
print(df['Sex'].value_counts())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', hue='Sex', data=df, palette='magma')
plt.title('Survival Count by Gender')
plt.xticks([0, 1], ['Died', 'Survived'])
plt.show()
```

Insight 1: The gender-based survival chart demonstrates that biological or physical strength was not the primary driver of survival. You will notice that while there were many more men on board, a much higher proportion of women survived.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.show()
```

Insight 2: The age distribution confirms that the Titanic was primarily populated by young adults in their prime working years, alongside a notable sub-group of young children. You'll see a large spike for infants (0-5 years) and young adults (20-35).
This demographic profile suggests a ship largely composed of families and individuals on board and in transition

```python
plt.figure(figsize=(10, 8))
numeric_df = df.select_dtypes(include=['number']) # Select only numbers
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Titanic Features')
plt.show()
```

Insight 3: The correlation heatmap reveals that socio-economic status (represented by Pclass and Fare) was the strongest numerical predictor of survival, while family size (SibSp/Parch) shows that many passengers traveled in groups; these family ties had a negligible impact on survival compared to the influence of ticket class and gender."
