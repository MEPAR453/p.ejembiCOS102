```python
# Python program to calculate Simple Interest
P = float(input("Enter principal: "))
R = float(input("Enter rate: "))
T = float(input("Enter time: "))
A = P * (1 + R/100) * T
print (A)
print("The Simple interest is", type(A))
print("The Simple interest is", A)
```

```python
# Python program to calculate Compound Interest
P = float(input("Enter principal: "))
R = float(input("Enter rate: "))
n = float(input("Enter number of compounding: "))
t = float(input("Enter time (in years): "))
A = P * (1 + R/n) ** (n*t)
print (A)
print("The Compound interest is", round( A, 2))
```

```python
# Python program to calculate Annuity Plan
PMT = float(input("Enter value of payment: "))
R = float(input("Enter rate: "))
n = float(input("Enter number of compounding: "))
t = float(input("Enter time (in years): "))
A = PMT * ((1 + R/n) ** (n*t) - 1) / (R/n)
print (A)
print("The annuity value is", round( A, 2))
```
